# Naive Bayes

```python
def prob_decision(df, decision):
    return len(df[df['Decision']==decision])/len(df)
```

```python
def joint_prob_nb(row, df, decision):

    cols = {0: 'Day', 1: 'Outlook', 2: 'Temperature', 3:'Humidity', 4:'Wind', 5:'Decision'}
    y_filtered_df = df[df['Decision']==decision]

    prob_x = 1
    prob_xy = 1
    
    for i in range(1,len(row)-1):
        x_i_val = row[i]
        #print(cols[i])
        #prob_x[i-1] = len(y_filtered_df[y_filtered_df[cols[i]==x_i_val]])/len(y_filtered_df)
        prob_xi = len(y_filtered_df[y_filtered_df[cols[i]]==x_i_val])/len(y_filtered_df)
        #prob_xi = len(y_filtered_df[y_filtered_df['Outlook']=='Sunny'])/len(y_filtered_df)
        prob_x = prob_x * prob_xi #len(y_filtered_df[y_filtered_df[cols[i]==x_i_val]])/len(y_filtered_df)
    
    prob_y = prob_decision(df, decision)

    prob_xy= prob_x * prob_y

    return prob_xy
```

```python
# Naive Bayes
prob_xy_yes = []
prob_xy_no = []
map_decision = []
for i,row in data.iterrows():
  #print(row)
  prob_xy_yes.append(joint_prob_nb(row,data,'Yes'))
  prob_xy_no.append(joint_prob_nb(row,data,'No'))
  map_est = 'Yes' if prob_xy_yes[i] > prob_xy_no[i] else 'No'
  map_decision.append(map_est)
```

```python
df_res = data.copy()
```

```python
df_res['prob_xy_yes'] = prob_xy_yes
df_res['prob_xy_no'] = prob_xy_no
df_res['MAP_est'] = map_decision
df_res.head()
```

```python
def calc_train_err(res_df):
    return len(res_df[res_df['Decision']!=res_df['MAP_est']])/len(res_df)
```

```python
calc_train_err(df_res)
```

# Laplace correction

```python
def joint_prob_nb_laplace(row, df, decision):

    cols = {0: 'Day', 1: 'Outlook', 2: 'Temperature', 3:'Humidity', 4:'Wind', 5:'Decision'}
    y_filtered_df = df[df['Decision']==decision]

    prob_x = 1
    prob_xy = 1

    laplace_factors = [3,3,2,2]
    
    for i in range(1,len(row)-1):
        x_i_val = row[i]
        #print(cols[i])
        #prob_x[i-1] = len(y_filtered_df[y_filtered_df[cols[i]==x_i_val]])/len(y_filtered_df)
        prob_xi = (len((y_filtered_df[y_filtered_df[cols[i]]==x_i_val]))+1)/(len(y_filtered_df)+laplace_factors[i-1])
        #prob_xi = len(y_filtered_df[y_filtered_df['Outlook']=='Sunny'])/len(y_filtered_df)
        prob_x = prob_x * prob_xi #len(y_filtered_df[y_filtered_df[cols[i]==x_i_val]])/len(y_filtered_df)
    
    prob_y = prob_decision(df, decision)

    prob_xy= prob_x * prob_y
    #print(prob_xy)
    return prob_xy
```

```python
prob_xy_yes_laplace = []
prob_xy_no_laplace = []
map_decision_laplace = []
for i,row in data.iterrows():
  #print(row)
  prob_xy_yes_laplace.append(joint_prob_nb_laplace(row,data,'Yes'))
  prob_xy_no_laplace.append(joint_prob_nb_laplace(row,data,'No'))
  map_est = 'Yes' if prob_xy_yes_laplace[i] > prob_xy_no_laplace[i] else 'No'
  map_decision_laplace.append(map_est)

  
```

```python
df_res_LC = data.copy()
```

```python
df_res_LC['prob_xy_yes_LC'] = prob_xy_yes_laplace
df_res_LC['prob_xy_no_LC'] = prob_xy_no_laplace
df_res_LC['MAP_est'] = map_decision_laplace
df_res_LC.head()
```

```python
calc_train_err(df_res_LC)
```
